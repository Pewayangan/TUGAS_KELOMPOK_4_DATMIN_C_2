```python
# display
%matplotlib inline  
# numerik
import numpy as np 
# analisis
import pandas as pd 
# plot
import matplotlib.pyplot as plt  
# visualisasi
import seaborn as sns
```

```python
#mengimport dataset
df1 = pd.read_csv("bapenda-od_15829_jml_kndrn__jenis_kndrn_fungsi_kndrn_cabang_pelayanan_data.csv")
df2 = pd.read_csv("bapenda-od_15831_jml_kndrn_belum_daftar_ulang__jenis_kndrn_fungsi_kndrn_data.csv")
df3 = pd.read_csv("bapenda-od_15832_jml_kndrn_tidak_daftar_ulang__jenis_kndrn_fungsi_kndrn_data.csv")
```

```python
df2.rename(columns={'jumlah_kendaraan': 'belum_daftar_ulang'}, inplace=True)
df3.rename(columns={'jumlah_kendaraan': 'tidak_daftar_ulang'}, inplace=True)
```

```python
df3
```

```python
#merge dataset dan dimasukkan kedalam variabel baru
df_EDA = pd.concat([df1, df2, df3], ignore_index=True)

# # Merge df1 dan df2 berdasarkan kolom yang dibutuhkan
result_1 = df1.merge(df2, on=['jenis_kendaraan', 'nama_kabupaten_kota', 'cabang_pelayanan', 'fungsi_kendaraan'], how='left')

# # Merge result_1 dengan df3
result_2 = result_1.merge(df3, on=['jenis_kendaraan', 'nama_kabupaten_kota', 'cabang_pelayanan', 'fungsi_kendaraan'], how='left')

# # result_2 berisi hasil gabungan dari ketiga DataFrame
result_2

#df_EDA
```

```python
#mengecek keberadaan data NULL disetiap kolom
result_2.isna().sum()
```

```python
hapus = [
    'id_y',
    'kode_provinsi_y',
    'nama_provinsi_y',
    'kode_kabupaten_kota_y',
    'satuan_y',
    'tahun_y',
    'id',
    'kode_provinsi',
    'nama_provinsi',
    'kode_kabupaten_kota',
    'satuan',
    'tahun'
]

# menghapus kolom-kolom tersebut dari DataFrame
result_2 = result_2.drop(columns=hapus)

result_2
```

```python
result_2.isna().head()
```

```python

# Mengelompokkan data dan menghitung jumlah kendaraan
result_df = result_2.groupby(['nama_kabupaten_kota', 'tahun_x'])['jumlah_kendaraan'].sum().reset_index()

# Membuat plot
plt.figure(figsize=(12, 6))
sns.barplot(data=result_df, x='nama_kabupaten_kota', y='jumlah_kendaraan', hue='tahun_x')

plt.xlabel('Nama Kabupaten/Kota')
plt.ylabel('Jumlah Kendaraan')
plt.title('Jumlah Kendaraan Berdasarkan tahun dan Nama Kabupaten/Kota Di Jawa Barat')
plt.legend(title='Fungsi Kendaraan')
plt.xticks(rotation=90)

# Mengatur format angka di sumbu Y agar tidak desimal
plt.gca().get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))

plt.show()
```

```python
# Mengelompokkan data dan menghitung jumlah kendaraan
result_df = result_2.groupby(['nama_kabupaten_kota'])['tidak_daftar_ulang'].sum().reset_index()

# Membuat plot
plt.figure(figsize=(12, 6))
sns.barplot(data=result_df, x='nama_kabupaten_kota', y='tidak_daftar_ulang')

plt.xlabel('Nama Kabupaten/Kota')
plt.ylabel('Jumlah Kendaraan')
plt.title('Jumlah Kendaraan Berdasarkan tahun dan Nama Kabupaten/Kota Di Jawa Barat')
plt.xticks(rotation=90)

# Mengatur format angka di sumbu Y agar tidak desimal
plt.gca().get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))

plt.show()
```

```python
# Mengelompokkan data dan menghitung jumlah kendaraan
result_df = result_2.groupby(['nama_kabupaten_kota'])['belum_daftar_ulang'].sum().reset_index()

# Membuat plot
plt.figure(figsize=(12, 6))
sns.barplot(data=result_df, x='nama_kabupaten_kota', y='belum_daftar_ulang')

plt.xlabel('Nama Kabupaten/Kota')
plt.ylabel('Jumlah Kendaraan')
plt.title('Jumlah Kendaraan Berdasarkan tahun dan Nama Kabupaten/Kota Di Jawa Barat')
plt.xticks(rotation=90)

# Mengatur format angka di sumbu Y agar tidak desimal
plt.gca().get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))

plt.show()
```
