```python
# display
%matplotlib inline  
# numerik
import numpy as np 
# analisis
import pandas as pd 
# plot
import matplotlib.pyplot as plt  
# visualisasi
import seaborn as sns
```

```python
#mengimport dataset
df1 = pd.read_csv("bapenda-od_15829_jml_kndrn__jenis_kndrn_fungsi_kndrn_cabang_pelayanan_data.csv")
df2 = pd.read_csv("bapenda-od_15831_jml_kndrn_belum_daftar_ulang__jenis_kndrn_fungsi_kndrn_data.csv")
df3 = pd.read_csv("bapenda-od_15832_jml_kndrn_tidak_daftar_ulang__jenis_kndrn_fungsi_kndrn_data.csv")
```

```python
#mengrename nama kolom 
df2.rename(columns={'jumlah_kendaraan': 'belum_daftar_ulang'}, inplace=True)
df3.rename(columns={'jumlah_kendaraan': 'tidak_daftar_ulang'}, inplace=True)
```

```python
#merge dataset dan dimasukkan kedalam variabel baru
df_EDA = pd.concat([df1, df2, df3], ignore_index=True)

# # Merge df1 dan df2 berdasarkan kolom yang dibutuhkan
result_1 = df1.merge(df2, on=['jenis_kendaraan', 'nama_kabupaten_kota', 'cabang_pelayanan', 'fungsi_kendaraan'], how='left')

# # Merge result_1 dengan df3
result_2 = result_1.merge(df3, on=['jenis_kendaraan', 'nama_kabupaten_kota', 'cabang_pelayanan', 'fungsi_kendaraan'], how='left')

# # result_2 berisi hasil gabungan dari ketiga DataFrame
result_2
```

```python
#mengecek keberadaan data NULL disetiap kolom
result_2.isna().sum()
```

```python
#memasukkan nama kolom yang tidak diperlukan kedalam variabel baru
hapus = [
    'id_y',
    'kode_provinsi_y',
    'nama_provinsi_y',
    'kode_kabupaten_kota_y',
    'satuan_y',
    'tahun_y',
    'id',
    'kode_provinsi',
    'nama_provinsi',
    'kode_kabupaten_kota',
    'satuan',
    'tahun'
]

# menghapus kolom-kolom tersebut dari DataFrame
result_2 = result_2.drop(columns=hapus)

result_2
```

```python
result_2.isna().sum()
```

```python
#mengrename nama kolom
result_2 = result_2.rename(columns={
    'id_x': 'id',
    'kode_provinsi_x': 'kode_provinsi',
    'nama_provinsi_x': 'nama_provinsi',
    'kode_kabupaten_kota_x': 'kode_kabupaten_kota',
    'satuan_x': 'satuan',
    'tahun_x': 'tahun'
})
result_2
```

```python
#mengecek inputan fungsi kendaraan yang didahului spasi
result_2.isin([' DINAS', ' UMUM', ' PRIBADI']).sum()
```

```python
#mengkoreksi input fungsi kendaraan yang didahului oleh spasi menjadi semua nya tanpa didahului spasi
result_2= result_2.replace(' DINAS', 'DINAS')
result_2= result_2.replace(' UMUM', 'UMUM')
result_2= result_2.replace(' PRIBADI', 'PRIBADI')
result_2.isin([' DINAS', ' UMUM', ' PRIBADI']).sum()
```

```python
# Mengelompokkan data dan menghitung jumlah kendaraan
result_df = result_2.groupby(['nama_kabupaten_kota', 'tahun'])['jumlah_kendaraan'].sum().reset_index()

# Membuat plot
plt.figure(figsize=(12, 6))
sns.barplot(data=result_df, x='nama_kabupaten_kota', y='jumlah_kendaraan', hue='tahun')

plt.xlabel('Nama Kabupaten/Kota')
plt.ylabel('Jumlah Kendaraan')
plt.title('Jumlah Kendaraan Berdasarkan tahun dan Nama Kabupaten/Kota Di Jawa Barat')
plt.legend(title='Fungsi Kendaraan')
plt.xticks(rotation=90)

# Mengatur format angka di sumbu Y agar tidak desimal
plt.gca().get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))

plt.show()
```

```python
sns.lineplot(x=result_2.tahun, y=result_2.jumlah_kendaraan, marker='o')

# menambahkan label
plt.xlabel('tahun_x')
plt.ylabel('jumlah_kendaraan')
plt.title('Perubahan jumlah kendaraan dari tahun 2013-2022')
```

```python
sns.lineplot(x=result_2.tahun, y=result_2.belum_daftar_ulang, marker='o')

# menambahkan label
plt.xlabel('tahun_x')
plt.ylabel('belum daftar ulang')
plt.title('Perubahan jumlah kendaraan yang belum daftar ulang dari tahun 2013-2022')
```

```python
sns.lineplot(x=result_2.tahun, y=result_2.tidak_daftar_ulang, marker='o')

# menambahkan label
plt.xlabel('tahun')
plt.ylabel('tidak daftar ulang')
plt.title('Perubahan jumlah kendaraan yang tidak daftar ulang dari tahun 2013-2022')
```

```python
# Menghitung rata-rata jumlah kendaraan tiap tahun
rata_rata_tiap_tahun = result_2.groupby('tahun')['jumlah_kendaraan'].mean()

# Membuat plot
plt.figure(figsize=(12, 6))
plt.plot(rata_rata_tiap_tahun.index, rata_rata_tiap_tahun.values, marker='o', linestyle='-')

plt.xlabel('Tahun')
plt.ylabel('Rata-Rata Jumlah Kendaraan')
plt.title('Rata-Rata Jumlah Kendaraan tiap Tahun')

plt.show()
```
